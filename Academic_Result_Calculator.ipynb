```python
# Welcome

print("Academic Result Calculator")
print("+-+-+-+-+-+-+-+-+-+-+-+-+-+")


# Name input

while True:
    name = input("Enter student name: ")
    if name.replace(" ", "").isalpha():
      break
    else:
        print("Name can consist of letters only.")


# Number of subjects

while True:
    num_subs_text = input("Enter number of subjects (max 10): ")
    if num_subs_text.isdigit():
        num_subs = int(num_subs_text)
        if 1 <= num_subs <= 10:
            break
        else:
            print("Max 10 subjects allowed.")
    else:
        print("Numbers only!")


# Here subject name + marks, will be saved

subjects = {}
total_marks = 0


# Loop
for i in range(1, num_subs+1):


  # Asking Subject name
  while True:
    sub_name = input(f"Enter subject {i} name: ")
    if sub_name.replace(" ", "").isalpha():
      break
    else:
      print("Name can consist of letters only.")


  # Asking Subject marks
  while True:
    sub_marks_text = input(f"Enter marks for {sub_name} (0-150): ")
    if sub_marks_text.isdigit():
      sub_marks = int(sub_marks_text)
      if 0 <= sub_marks <= 150:
        subjects[sub_name] = sub_marks
        total_marks += sub_marks
        break
      else:
        print("Marks should be between 0 and 150.")
    else:
      print("Numbers only!")


# Percentage
max_total = num_subs * 150
percentage_raw = (total_marks / max_total) * 100
percentage = round(percentage_raw)

# Grades
if percentage >= 95:
  grade = "A*"
elif percentage >= 81:
  grade = "A"
elif percentage >= 71:
  grade = "B"
elif percentage >= 61:
  grade = "C"
elif percentage >= 51:
  grade = "D"
else:
  grade = "F"


# Final(Result) Output
print("RESULT CARD")
print("=-=-=-=-=-=-=")
print("Name:", name)

print("Total:", total_marks, "/", i * 150)
print("Percentage:", percentage)
print("Grade:", grade)
```
