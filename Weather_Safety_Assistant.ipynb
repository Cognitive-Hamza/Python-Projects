```python
name = input("Enter your name: ")

if name == "":
	print("You did not enter name, please run again.")
else:
	temp_text = input("Enter temperature outside: ")

	if temp_text.replace("-", "").isdigit():
		temp_text = int(temp_text)

		if temp_text >= 30:
			print("It's hot! Stay hydrated.")
		elif temp_text >= 15:
			print("Nice weather! You can play outside.")
		elif temp_text >= 0:
			print("Its's cold wear a jacket.")
		else:
			print("Too cold! Stay indoors.")
	else:
		print("That is not a valid number. Please enter digits only next time.")
```
