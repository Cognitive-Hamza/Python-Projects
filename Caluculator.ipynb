```python
# Welcome the user
print("Simple Calculator")
print("=-=-=-=-=-=-=-=-=-=")

welcome = "Welcome to the Calculator!"
print(welcome)

# Ask user for numbers (with validation)

num1_text = input("Enter the first number: ")
if num1_text == "":
    print("Error: Empty input is not allowed.")
    exit()
else:
    num2_text = input("Enter the second number: ")
    if num2_text == "":
        print("Error: Empty input is not allowed.")
        exit()
    else:

        # Check for non-numeric input
        if num1_text.isdigit():
            if num2_text.isdigit():

                # Convert to numbers
                num1 = float(num1_text)
                num2 = float(num2_text)

                # Give the user operations' options
                print("Press 1 for addition '+'")
                print("Press 2 for subtraction '-'")
                print("Press 3 for multiplication '*'")
                print("Press 4 for division '/'")

                # Ask for the function
                operator = input("Enter the required operation: ")

                # Perform calculation
                if operator == "1":
                    print(num1 + num2)

                elif operator == "2":
                    print(num1 - num2)

                elif operator == "3":
                    print(num1 * num2)

                elif operator == "4":
                    if num2 == 0:
                        print("Error: Cannot divide by zero!")
                    else:
                        print(num1 / num2)

                else:
                    print("Invalid operator selected.")

            else:
                print("Error: Please enter numbers only.")
                exit()
        else:
            print("Error: Please enter numbers only.")
            exit()
```
