```python
# Pakistan Electricity Bill (Very Simple Version)
# Rules:
# - First 200 units: Rs 18 per unit
# - Above 200 units: Rs 45 per unit
# - Fixed meter charge: Rs 150 for every unit

RATE_LOW = 18		# Rs per unit for first 200 units
RATE_HIGH = 45		# Rs per unit for units above 200
FIXED_CHARGE = 150	# Rs

print ("Welcome to the Electricity Bill Calculator (Pakistan)")
name = input("Enter your name: ")

# Ask the student for number of units (whole number)
units_text = input("Enter the number of units used: ")
units = int(units_text) # change text to number

# Check units and calculate bill
if units <= 200:
	# All units are in the low rate
	energy_cost = units * RATE_LOW
	bill_category = "Low (<= 200 units)"
else:
	# First 200 units at low rate, the rest at high rate
	first_part = 200 * RATE_LOW
	extra_units = units - 200
	second_part = extra_units * RATE_HIGH
	energy_cost = first_part + second_part
	bill_category = "High (>200 units)"

total_bill = energy_cost + FIXED_CHARGE

# Show the bill nicely
print("\n--- Electricity Bill ---")
print("Name:", name)
print("Units:", units)
print("Bill Category", bill_category)
print("Fixed Charge (Rs):", FIXED_CHARGE)
print("Total Bill (Rs):", total_bill)
print("------------------------")
```
