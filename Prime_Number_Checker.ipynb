```python
# Prime number checker

welcome = "Welcome to the PRIME NUMBER checker!"
print(welcome)

name = input("Enter your name: ")

# Check for empty name

if name == "":
	print("You did not type your name. Please run the program again and type something.")
	exit()

# Check for digits in name

elif any(char.isdigit() for char in name):
	print("Invalid name(numbers are not allowed). Please run the program again.")
	exit()

# Valid name

else:
	print("Hello,", name + "!")

# Take input for number to check

num_text = input("Enter a number: ")

# Check if number is valid

if num_text.isdigit():  # checks if input is digits only
	num = int(num_text)  # convert text to number

	if num > 1:
		for i in range(2, num):
			if num % i == 0:
				print(f"{name}, this is NOT a PRIME number.")
				break
		else:
            		print(f"{name}, this IS a PRIME number!")
	else:
		print(f"{name}, this is NOT a PRIME number.")
else:
	print("This does not look like a number. Please enter digits only.")
```
