```python
print("Quiz Game + Score History")
print("------------------------------------")
print("This program allows you to search results, run a quiz, and save scores.\n")

results = []
names_set = set()

# Menu
def show_menu():
    print("\n+++ Quiz Menu +++")
    print("0 - Exit")
    print("1 - Start Quiz")
    print("2 - View all results")
    print("3 - Search results by name")
    print("4 - Save results in a file")

# Initiate Quiz
def start_quiz():
    name = input("Enter your name: ").strip()
    if name == "":
        print("Name cannot be empty!")
        return

    score = 0

    questions = [
        ("What is 2 - 2? ", "0"),
        ("What is 5 + 13? ", "18"),
        ("What is 10 * 6? ", "60"),
        ("What is 3 / 3? ", "1"),
        ("What is the capital city of Pakistan? ", "islamabad")
    ]

    print("\nAnswer the questions:")
    for i in range(len(questions)):
        answer = input(f"Q{i+1}: {questions[i][0]}").strip().lower()

        if answer == questions[i][1]:
            print("Correct!")
            score += 1
        else:
            print("Wrong. Correct answer is:", questions[i][1].title())

    # Feedback
    print("\nFinal Score:", score, "/ 5")
    match score:
        case 5:
            print("Excellent! Perfect.")
        case 4:
            print("Great Job! Almost perfect.")
        case 3:
            print("Good Effort! Keep practicing.")
        case 2:
            print("You have the potential! Practice more.")
        case _:
            print("Don't worry. Try again.")

    record = f"{name} - {score}/5"
    results.append(record)
    names_set.add(name.lower())
    print("Saved result:", record)

# View all results
def view_results():
    if len(results) == 0:
        print("No results saved yet.")
        return

    print("\nAll Saved Results:")
    for i, r in enumerate(results, start=1):
        print(f"{i}. {r}")

# Search by name
def search_results():
    if len(results) == 0:
        print("No results to search.")
        return

    search_name = input("Enter name to search: ").strip().lower()
    if search_name == "":
        print("Search name cannot be empty.")
        return

    if search_name not in names_set:
        print("No results found for that name.")
        return

    print("\nResults for:", search_name)
    found = 0
    for r in results:
        if r.lower().startswith(search_name + " -"):
            print("-", r)
            found += 1

    if found == 0:
        print("No results found.")

# Save results to file
def save_to_file():
    if len(results) == 0:
        print("No results to save.")
        return

    filename = "quiz_results.txt"
    with open(filename, "w") as f:
        f.write("Quiz Results History\n")
        f.write("--------------------\n")
        for r in results:
            f.write(r + "\n")

    print("Results saved to file:", filename)

# Main loop
while True:
    show_menu()
    option_text = input("Choose an option (0-4): ").strip()

    if not option_text.isdigit():
        print("Invalid input. Please enter a number from 0 to 4.")
        continue

    option = int(option_text)

    match option:
        case 1:
            start_quiz()
        case 2:
            view_results()
        case 3:
            search_results()
        case 4:
            save_to_file()
        case 0:
            print("Goodbye!")
            break
        case _:
            print("Invalid option. Choose between 0 and 4.")

print("\nProgram finished.")
```
