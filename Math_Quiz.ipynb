```python
print("Simple Math Quiz Program")
print("*"*25)
print("This program asks 3 basic math questions and gives feedback")

# Get student's name
name = input("What is your name? ")
print(f"Hello {name}! Let's start the math quiz.\n")

# Initialize score
score = 0

# Question 1
print("Question 1: What is 5 + 3?")
answer1 = input("Your answer: ")
if answer1 == "8":
    print("Correct!\n")
    score = score + 1
else:
    print("Wrong! The correct answer is 8.\n")

# Question 2
print("Question 2: What is 10 - 4?")
answer2 = input("Your answer: ")
if answer2 == "6":
    print("Correct!\n")
    score = score + 1
else:
    print("Wrong! The correct answer is 6.\n")

# Question 3
print("Question 3: What is 3 * 4?")
answer3 = input("Your answer: ")
if answer3 == "12":
    print("Correct!\n")
    score = score + 1
else:
    print("Wrong! The correct answer is 12.\n")

# Show final score
print(f"Quiz completed, {name}!")
print(f"Your final score: {score} out of 3")

# Give feedback based on score
if score == 3:
    print("Excellent! You got all questions correct!")
elif score == 2:
    print("Good job! You got most questions correct.")
elif score == 1:
    print("Keep practicing! You can do better.")
else:
    print("Don't worry! Practice makes perfect.")
```
