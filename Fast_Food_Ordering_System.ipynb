Requirements:

Fast Food Ordering System

Introduction
This project is about a Fast Food Service System that displays a welcome message, shows a
complete menu with prices, and ensures that the user enters only valid input. The system is
designed for a small fast-food outlet offering items like burgers, broast, club sandwiches,
and tikka.

Objectives
1. To display a welcome message for customers.
2. Asks the user for their name
3. To show a fast-food menu clearly.
4. To provide correct prices of all food items.
5. To prevent users from entering special characters or extremely large values.
6. To make the ordering experience simple and user-friendly.
7. Limit of ordering any items more than 10 in quantity

Menu of the Outlet:
Zinger Burger – Rs 400,
Chest Broast – Rs 550,
Leg Broast - 500,
Club Sandwich – Rs 450,
Tikka – Rs 300,
10 Nuggets - Rs 350,
Fries - Rs 100,
Fizup - Rs 70,
Cola Next - Rs 70,
Fresh Juice - 120

Input Validation / Robustness
To keep the system stable and user-friendly, the following rules are applied:
- No special characters allowed (@, #, $, %, &amp;, *)
- No letters allowed (only numbers are accepted)
- No extremely large numbers allowed
- Only menu-related numbers (1–4) are accepted
If a user gives wrong input, the system shows an error and asks again.

Working of the System
1. Displays a welcome message.
2. Shows the complete fast-food menu.
3. Shows Bill

```python
print("Fast Food Ordering System")
print("-------------------------")

grand_total = 0

print("\nMenu:")
print("1 - Zinger Burger (Rs 400)")
print("2 - Chest Broast (Rs 550)")
print("3 - Leg Broast (Rs 500)")
print("4 - Club Sandwich (Rs 450)")
print("5 - Tikka (Rs 300)")
print("6 - Fries (Rs 100)")
print("7 - 10 Nuggets (Rs 350)")
print("8 - Cola Next (Rs 70)")
print("9 - Fizup (Rs 70)")
print("10 - Frech Juice (Rs 120)")
print("0 - Finish Order")

print("\nYou can order up to 5 items.")

for order in range(5):

    choice_text = input("\nEnter item number (0–4): ")

    if not choice_text.isdigit():
        print("Invalid input! Numbers only.")
        continue

    choice = int(choice_text)

    if choice == 0:
        break

    if choice < 1 or choice > 10:
        print("Invalid menu choice!")
        continue

    qty_text = input("Enter quantity: ")

    if not qty_text.isdigit():
        print("Invalid quantity!")
        continue

    quantity = int(qty_text)

    if quantity <= 0:
        print("Quantity must be at least 1.")
        continue

    match choice:
        case 1:
            item = "Burger"
            price = 400
        case 2:
            item = "Broast"
            price = 550
        case 3:
            item = "Club Sandwich"
            price = 450
        case 4:
            item = "Tikka"
            price = 300

    total = price * quantity
    grand_total += total

    print(f"Added: {item} x {quantity} = Rs {total}")

print("\n================ RECEIPT ================")
print("Total Bill: Rs", grand_total)
print("Thank you for ordering!")
print("=========================================")
```

```python
# Fast Food Ordering System

# Program title
print("Fast Food Ordering System")
print("=-=-=-=-=-=-=-=-=-=-=-=-=")


name = input("Please enter your name: ")

# Name validation:
# 1. name != "" → empty name not allowed
# 2. replace(" ", "") → removes spaces
# 3. Allows first name + last name
# 4. isalpha() → checks only letters
# 5. Rejects numbers, special chars, empty input

if name != "" and name.replace(" ", "").isalpha():

    # Welcome message
    print(f"\nWelcome, {name}!")

    # Displaying menu

    print("\nMenu:")
    print("0 - Finish Order")
    print("1 - Tikka (Rs 300)")
    print("2 - Burger (Rs 400)")
    print("3 - Club Sandwich (Rs 450)")
    print("4 - Broast (Rs 550)")

    # Order rules
    print("\nYou can order up to 10 of each item (TOTAL).")


    # Variable to store final bill

    grand_total = 0

    # Variables to store TOTAL quantity
    # of each food item

    tikka_total = 0
    burger_total = 0
    club_total = 0
    broast_total = 0

    # Loop allows user to order
    # maximum 10 times

    for order in range(10):

        # Taking menu choice
        choice_text = input("\nEnter item number (0–4): ")

        # Checking if input is numeric
        if choice_text.isdigit():

            choice = int(choice_text)

            # If user chooses 0, break
            if choice == 0:
                break

            # Valid menu choice check
            if choice >= 1 and choice <= 4:

                # Taking quantity
                qty_text = input("Enter quantity: ")

                # Quantity must be numeric
                if qty_text.isdigit():

                    quantity = int(qty_text)

                    # Quantity range check
                    if quantity >= 1 and quantity <= 10:

                        # Tikka ordering logic
                        if choice == 1:
                            if tikka_total + quantity <= 10:
                                tikka_total += quantity
                                total = quantity * 300
                                grand_total += total
                                print(f"Added: Tikka x {quantity} = Rs {total}")
                            else:
                                print("You cannot order more than 10 Tikka in total.")

                        # Burger ordering logic
                        elif choice == 2:
                            if burger_total + quantity <= 10:
                                burger_total += quantity
                                total = quantity * 400
                                grand_total += total
                                print(f"Added: Burger x {quantity} = Rs {total}")
                            else:
                                print("You cannot order more than 10 Burger in total.")

                        # Club Sandwich ordering logic
                        elif choice == 3:
                            if club_total + quantity <= 10:
                                club_total += quantity
                                total = quantity * 450
                                grand_total += total
                                print(f"Added: Club Sandwich x {quantity} = Rs {total}")
                            else:
                                print("You cannot order more than 10 Club Sandwich in total.")

                        # Broast ordering logic
                        elif choice == 4:
                            if broast_total + quantity <= 10:
                                broast_total += quantity
                                total = quantity * 550
                                grand_total += total
                                print(f"Added: Broast x {quantity} = Rs {total}")
                            else:
                                print("You cannot order more than 10 Broast in total.")

                    else:
                        # Quantity outside allowed range
                        print("Quantity must be between 1 and 10.")

                else:
                    # Quantity contains non-numeric characters
                    print("Invalid quantity! Numbers only.")

            else:
                # Menu choice not between 1 and 4
                print("Invalid menu choice!")

        else:
            # Menu input is not numeric
            print("Invalid input! Numbers only.")

    # Printing receipt

    print("\n=_=_=_=_=_=_=_= RECEIPT =_=_=_=_=_=_=_=_=")
    print("Customer Name:")
    print(name)
    print("---------------------------------------")
    print("Items ordered:")

    # Printing only items that were ordered

    if tikka_total > 0:
        print(f"Tikka x {tikka_total} = Rs {tikka_total * 300}")
    if burger_total > 0:
        print(f"Burger x {burger_total} = Rs {burger_total * 400}")
    if club_total > 0:
        print(f"Club Sandwich x {club_total} = Rs {club_total * 450}")
    if broast_total > 0:
        print(f"Broast x {broast_total} = Rs {broast_total * 550}")

    # Final bill

    print("---------------------------------------")
    print("Total Bill: Rs", grand_total)
    print("Thank you for ordering!")
    print("-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-")

else:
    # Invalid name entered (Showing error)
    print("Invalid name! Name must contain letters only and cannot be empty.")
```
