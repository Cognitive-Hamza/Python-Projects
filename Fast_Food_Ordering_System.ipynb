Requirements:

Fast Food Ordering System

Introduction
This project is about a Fast Food Service System that displays a welcome message, shows a
complete menu with prices, and ensures that the user enters only valid input. The system is
designed for a small fast-food outlet offering items like burgers, broast, club sandwiches,
and tikka.

Objectives
1. To display a welcome message for customers.
2. Asks the user for their name
3. To show a fast-food menu clearly.
4. To provide correct prices of all food items.
5. To prevent users from entering special characters or extremely large values.
6. To make the ordering experience simple and user-friendly.
7. Limit of ordering any items more than 3 in quantity

Menu of the Outlet
Burger – Rs 400
Broast – Rs 550
Club Sandwich – Rs 450
Tikka – Rs 300

Input Validation / Robustness
To keep the system stable and user-friendly, the following rules are applied:
- No special characters allowed (@, #, $, %, &amp;, *)
- No letters allowed (only numbers are accepted)
- No extremely large numbers allowed
- Only menu-related numbers (1–4) are accepted
If a user gives wrong input, the system shows an error and asks again.

Working of the System
1. Displays a welcome message.
2. Shows the complete fast-food menu.
3. Shows Bill

## Testing and prototype of code

```python
print("Fast Food Ordering System")
print("-------------------------")

grand_total = 0

print("\nMenu:")
print("1 - Burger (Rs 400)")
print("2 - Broast (Rs 550)")
print("3 - Club Sandwich (Rs 450)")
print("4 - Tikka (Rs 300)")
print("0 - Finish Order")

print("\nYou can order up to 5 items.")

for order in range(5):

    choice_text = input("\nEnter item number (0–4): ")

    if not choice_text.isdigit():
        print("Invalid input! Numbers only.")
        continue

    choice = int(choice_text)

    if choice == 0:
        break

    if choice < 1 or choice > 4:
        print("Invalid menu choice!")
        continue

    qty_text = input("Enter quantity: ")

    if not qty_text.isdigit():
        print("Invalid quantity!")
        continue

    quantity = int(qty_text)

    if quantity <= 0:
        print("Quantity must be at least 1.")
        continue

    match choice:
        case 1:
            item = "Burger"
            price = 400
        case 2:
            item = "Broast"
            price = 550
        case 3:
            item = "Club Sandwich"
            price = 450
        case 4:
            item = "Tikka"
            price = 300

    total = price * quantity
    grand_total += total

    print(f"Added: {item} x {quantity} = Rs {total}")

print("\n================ RECEIPT ================")
print("Total Bill: Rs", grand_total)
print("Thank you for ordering!")
print("=========================================")
```
