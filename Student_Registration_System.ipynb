```python
print("===== Student Registration System =====")

name = input("Student name: ")
age = input("Student age: ")
roll_no = input("Roll number: ")

# Validation
if name.isalpha():

    if age.isdigit():

        if roll_no.isdigit():

            # Save data
            f = open("students.txt", "a")
            f.write("Name: " + name + "\n")
            f.write("Age: " + age + "\n")
            f.write("Roll No: " + roll_no + "\n")
            f.write("----------------------\n")
            f.close()

            print("\nâœ… Registration Successful!")
            print("\nðŸ“‚ Saved Students Data:\n")

            # Read and show data
            f = open("students.txt", "r")
            data = f.read()
            print(data)
            f.close()

        else:
            print("Roll number should be numbers only.")

    else:
        print("Age should be numbers only.")

else:
    print("Name should be aphabets only.")
```
