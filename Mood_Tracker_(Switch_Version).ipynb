```python
# Welcome

print("Mood Tracker (Switch Version)")
print("+_+_+_+_+_+_+_+_+_+_+_+_+_+_+")

# Ask for name of the user

name = input("Please enter your name: ")

# Checking if name is empty

if name == "":
	print("Name cannot be left empty. Re-run the program please.")

# Continuing the main program

else:
	print("\nHello", name, "! You will enter your mood 3 times.")
	print("\nMood Choices:")
	print("1 = Sad")
	print("2 = Okay")
	print("3 = Happy")

	# Counters of the moods' repetitveness
	sad_mood = 0
	okay_mood = 0
	happy_mood = 0

	# Function for displaying message according to mood using 'match'

	def mood_message(mood_number):
		match mood_number:
			case 1:
				print("Try talking to someone you trust.")
			case 2:
				print("Keep going and stay positive!")
			case 3:
				print("Great! Keep smiling and sharing kindness.")
			case _:
				print("Invalid mood number. Please enter 1, 2, or 3.")

	# For loop

	for day in range(1, 4):
		print("\nDay", day)
		mood_text = input("Enter your mood (1â€“3): ")

		if mood_text.isdigit():
			mood = int(mood_text)

			# Use switch (match) for message

			mood_message(mood)

			# Count moods for summary

			match mood:
				case 1:
					sad_mood += 1
				case 2:
					okay_mood += 1
				case 3:
					happy_mood += 1
				case _:
					print("Invalid input. Enter input again")
		else:
			print("Invalid input. Please enter numbers only (1, 2, or 3).")

	# Use the mood sums to get the mood the user is in most of the time

	if happy_mood == okay_mood == sad_mood:
		print("Your mood was mostly okay. Keep building good habits.")
	elif happy_mood >= sad_mood and happy_mood >= okay_mood:
		print("Nice! You were happy most of the time.")
	elif okay_mood >= sad_mood:
		print("Your mood was mostly okay. Keep building good habits.")
	else:
		print("You had more sad days. Talk to someone you trust.")

	# Final Summary

	print("\nMood Summary for", name)
	print("Sad days: ", sad_mood)
	print("Okay days: ", okay_mood)
	print("Happy days: ", happy_mood)
```
