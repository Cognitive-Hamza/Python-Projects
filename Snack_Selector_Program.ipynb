```python
'''Display Title'''

print("Snack Selector Program")
print("----------------------")


# We want the user to choose a snack up to 3 times. So we add a for loop.

for attempt in range(3):  # range(3) will allow the loop to run 3 times.
	print("\nChoose a snack from the menu: ")

# Show the menu inside the loop. Inside the loop (indented), add:
	print("0 - Exit")
	print("1 - Apple")
	print("2 - Banana")
	print("3 - Sandwich")
	print("4 - Juice")

# Ask the user for a choice. While still staying inside the loop.
	choice_str = input("Type a number (0-4): ")
	choice = int(choice_str)

# We are trusting them to type a number for this simple program, that is why will not include any checks.
	if choice == 0:
		print("Okay, exiting. Bye!")
		break

# Now we use Pythonâ€™s match (like switch) to decide what to print based on the snack choice.
	match choice:
		case 1:
			print("You chose an Apple. Healthy choice!")
		case 2:
			print("You chose a Banana. Great for energy!")
		case 3:
			print("You chose a Sandwich. Yummy!")
		case 4:
			print("You chose Juice. Stay refreshed!")
		case _:
			print("That number does not represent anything in the menu.")
```
