```python
import sys

# Welcome
print("Homework Manager")
print("|!|!|!|!|!|!|!|!|")
print("This program lets you add, view, search, and save homework tasks.")

# Data
subjects = set()
tasks = []

# Menu
def show_menu():
	print("\nMENU:")
	print("0 - Exit")
	print("1 - Add Homework")
	print("2 - View All Homework")
	print("3 - Search Homework According to Subject")
	print("4 - Save Homework In a File")

# Add homework
def add_hw():
	subject = input("Enter the subject name: ").strip()
	if subject == "":
		print("Subject name cannot be empty.")
		return

	task = input("Enter the task: ").strip()
	if task == "":
		print("Task cannot be empty.")
		return

	complete_task = f"{subject} → {task}"
	tasks.append(complete_task)
	subjects.add(subject.lower())

	print("Saved:", complete_task)

# View homework
def view_tasks():
	if not tasks:
		print("No homework saved.")
		return

	print("\nAll Tasks:")
	for i, task in enumerate(tasks, start=1):
		print(f"{i}. {task}")

# Search homework
def search_by_subject():
	if not tasks:
		print("No homework saved yet.")
		return

	search = input("Enter subject to search: ").strip().lower()
	if search == "":
		print("Search subject cannot be empty.")
		return

	found = False
	for t in tasks:
		if t.lower().startswith(search + " →"):
			if not found:
				print(f"\nTasks for subject '{search}':")
			print("-", t)
			found = True

	if not found:
		print("No tasks found for that subject.")

# Save to file
def save_to_file():
	if not tasks:
		print("No homework to save.")
		return

	with open("homework_tasks.txt", "w") as f:
		for t in tasks:
			f.write(t + "\n")

	print("Homework saved to homework_tasks.txt")

# Exit program
def end():
	print("Goodbye!")
	sys.exit()

# Main loop
while True:
	show_menu()
	choice_text = input("Choose an option (0-4): ").strip()

	if not choice_text.isdigit():
		print("Invalid input. Enter a number.")
		continue

	choice = int(choice_text)

	if choice == 0:
		end()
	elif choice == 1:
		add_hw()
	elif choice == 2:
		view_tasks()
	elif choice == 3:
		search_by_subject()
	elif choice == 4:
		save_to_file()
	else:
		print("Invalid choice. Choose between 0 and 4.")
```
