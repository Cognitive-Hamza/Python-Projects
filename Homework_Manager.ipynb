```python
# Welcome

print("Homework Manager")
print("|!|!|!|!|!|!|!|!|")
print("This program lets you add, view, search, and save homework tasks.")


# Counter:

subjects: set()
tasks: []


# Show menu

def show_menu():
	print("\nMENU:")
	print("1 - Add Homework.")
	print("2 - View All Homework.")
	print("3 - Search Homework According to Subject.")
	print("4 - Sace Homework In a File.")


# Add homework

def add_hw():
	subject = input("Enter the subject name: ")
	if subject == ""
		print("Subject name cannot be empty.")
		return

	task = input("Enter the task of the subject you entered: ")
	if task == ""
		print("Task cannot be empty")
		return

	complete_task: subject, "â†’", task
	tasks.append(complete_task)
	subjects.add(subject.lower())

	print("Saved: ", complete_task)


# View all tasks

def view_tasks():
	if len(tasks) == 0:
		print("None saved.")
		return

	print("All Tasks: ")
	for attempt in range(len(tasks)):
		print(str(i + 1) + ".", tasks[i])


# Searching

def search_by_subject():
	if len(tasks) == 0:
		print("No homework saved yet.")
		return

	search = input("Enter subject to search: ").strip().lower()
	if search == "":
		print("Search subject cannot be empty.")
		return

	if search not in subjects:
		print("No tasks found for that subject.")
		return

	print("\nTasks for subject:", search)
	found = 0
	for t in tasks:
		if t.lower().startswith(search + " -"):
			print("-", t)
			found = found + 1

	if found == 0:
		print("No tasks found (unexpected case).")


# Saving

def save_to_file():
	if len(tasks) == 0:
		print("No homework to save.")
		return

	filename = "homework_tasks.txt"
	f = open(filename, "w")
	for t in tasks:
		f.write(t + "\n")
	f.close()
	print("Homework saved to file:", filename)


# Main loop (menu repeats)

while True:
	show_menu()
	choice_text = input("Choose an option (0-4): ").strip()

	if choice_text.isdigit():
		choice = int(choice_text)
	else:
		print("Invalid input. Please enter a number from 0 to 4.")
		continue

	match choice:
		case 1:
			add_homework()
		case 2:
			view_homework()
		case 3:
			search_by_subject()
		case 4:
			save_to_file()
		case 0:
			print("Goodbye!")
			break
		case _:
			print("Invalid choice. Please choose 0, 1, 2, 3, or 4.")

print("\nProgram finished.")
```
