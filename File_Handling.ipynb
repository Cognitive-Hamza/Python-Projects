```python
# Asks the user for their name and age.

name = input("Enter your name: ")
age_text = input("Enter your age: ")
age = int(age_text)

# Saves this data to users.txt (one line per user).

with open("user.txt", "w") as f:
	f.write(f"You name is {name} \n")
	f.write(f"You are {age} years old \n")

# After writing, opens the same file and prints all lines.

with open("user.txt", "r") as f:
	content = f.read()
	print(content)
```
