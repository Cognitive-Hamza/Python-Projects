```python
print("Pocket Money Tracker")
print("=-=-=-=-=-=-=-=-=-=-=")

name = input("Enter your name: ")

if name == "":
	print("Please enter your name.")
else:

	pocket_money_text = input("Enter your daily pocket money: ")

	if pocket_money_text.isdigit():
		pocket_money = int(pocket_money_text)
		if pocket_money <= 200:
			total_spent = 0

			def spending_advice(saved_up_money):
				if saved_up_money >= 100:
					print("Excellent! You saved a great amount of your pocket money.")
				elif saved_up_money >= 50:
					print("Great job! You saved a lot.")
				elif saved_up_money >= 20:
					print("Good! Try to save more next time.")
				else:
					print("Be careful! You spent almost all of your money.")

			for item in range(1, 4):
				print("\nItem", item)
				spend_text = input("Enter money spent: ")

				if spend_text.isdigit():
					spent = int(spend_text)
					total_spent = total_spent + spent
				else:
					print("Invalid input. Money spent set to 0.")

			saved_up_money = pocket_money - total_spent
			if saved_up_money < 0:
				print("You definitely messed up the inputs. Please re-run porgram.")
				print("Spent amount is greater than you initial amount.")
				exit()
			else:

				print("Final Tracking Result for ", name)
				print("Total Pocket Money:", pocket_money)
				print("Total Spent:", total_spent)
				print("Money Saved:", saved_up_money)

				spending_advice(saved_up_money)

				print("Thank you for using Pocket Money Tracker.")
		else:
			print("Above 200 is clearly unrealistic. Please input a valid amount.")
	else:
		print("Invalid amount. Please enter numbers only.")
```
