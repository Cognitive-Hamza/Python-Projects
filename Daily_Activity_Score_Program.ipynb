```python
# Daily Activity Score Program

# Show Title

print("Daily Activity Score Program")
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")

total_time = 0
day_number = 1

# Ask user for thier name

name = input("Enter your name please: ")

# Welcome

if name == "":
	print("Please enter your name first.")
	exit()
else:
	print("Hello, " + name + " please enter your study time.")

# function

	def study_time_feedback(time):
		if time >= 720:
			print("That is clearly unrealistic")
		elif time >= 60:
			print("Excellent Work!")
		elif time >= 45:
			print("Good job!")
		elif time >= 0:
			print("Try to study more.")
		else:
			print("That is not a valid time period.")
			exit()


	for day_number in range(1, 4):
		print(f"Day: ", {day_number})
		time_text = input("Enter study time in minutes: ")


		if time_text.replace("-", "").isdigit():
			time = int(time_text)
			study_time_feedback(time)
			day_number += 1
		else:
			print("Enter numbers only.")
			break

	print("Total study time for all days:", total_time, "minutes")
	print("Allah Hafiz " + name + "! Keep it up.")
```
