```python
name = input("Enter your name: ")

if name == "":
	print("You did not enter a name. Please run again.")
else:
	score_text = input("Enter your score (0-100): ")

	if score_text.isdigit():
		score = int(score_text)

		if score < 0 or score > 100:
			print("Score should be between 0 and 100.")
		else:
			if score >= 80:
				print("Excellent, ", name + "!")
			elif score >= 50:
				print("Good job, ", name + "!")
			else:
				print("Keep trying, ", name + "!")
	else:
		print("That is not a valid number. Please enter digits only next time.")
```
