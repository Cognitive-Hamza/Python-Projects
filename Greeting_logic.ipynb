```python
# Greeting Program
# This program asks for a student's name and age, then prints a friendly message

# Get student's name
name = input("What is your name: ")

# Get student's age and validate it's a number
age = input("What is your age: ")

# Check if age is a valid number
if age.isdigit():
    age = int(age)
    print(f"\nWelcome {name}! It's great to have a {age}-year-old student here!")
else:
    print(f"\nError: Invalid age entered for {name}. Please enter a numerical value for age.")
```
