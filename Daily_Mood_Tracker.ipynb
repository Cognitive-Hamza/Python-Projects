```python
def get_supportive_message(name, mood_choice):

	if mood_choice == 1:
		print(f"\n{name}, I'm sorry you're feeling sad. It's okay to feel this way. You're not alone.")
	elif mood_choice == 2:
		print(f"\n{name}, thanks for sharing. I hope your day gets even better.")
	elif mood_choice == 3:
		print(f"\n{name}, that's great to hear! Keep enjoying the positive moments.")
	elif mood_choice == 0:
		exit()
	else:
		print(f"\n{name}, please enter a valid option (1, 2, or 3).")


# Main program
print("Welcome to the Daily Mood Tracker")
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")

name = input("What is your name? ")

if name != "" and name.replace(" ", "").isalpha():

	print("\nHow are you feeling today?")
	print("1 = Sad")
	print("2 = Okay")
	print("3 = Happy")

	mood_text = input("Please enter 1, 2, or 3: ")

	if mood_text.isdigit():
		mood_choice = int(mood_text)
		get_supportive_message(name, mood_choice)
	else:
		print("Enter numbers only (1-3).")

else:
	print("Please enter a valid name.")
```
