```python
print("Student Mini-Shop")
print("----------------------------")
print("Shop within a budget, select items, and generate a receipt file.\n")

# storing items in the format: "Item - Cost"
shopping_cart = []
# keep track of different items bought
items_bought = set()
# initialize budget
customer_budget = -1


# Display the main options menu
def display_menu():
    print("\nMAIN MENU")
    print("1 - Enter Budget")
    print("2 - Select Items (3 max)")
    print("3 - Show Cart and Total")
    print("4 - Finalize Purchase & Save Receipt")
    print("0 - Exit Program")


def display_shop_items():
# Displays available items in the shop
    print("\nAVAILABLE ITEMS")
    print("1 - Pencil (10)")
    print("2 - Eraser (5)")
    print("3 - Notebook (25)")
    print("4 - Juice (30)")
    print("5 - Sandwich (50)")


def fetch_item_details(selection):
# Returns item name and price based on user selection
    match selection:
        case 1:
            return ("Pencil", 10)
        case 2:
            return ("Eraser", 5)
        case 3:
            return ("Notebook", 25)
        case 4:
            return ("Juice", 30)
        case 5:
            return ("Sandwich", 50)
        case _:
            return ("", -1)


def calculate_cart_sum():
# Calculates the total cost of all items in the cart
    total_cost = 0
    for item in shopping_cart:
        parts = item.split(" - ")
        if len(parts) == 2 and parts[1].isdigit():
            total_cost += int(parts[1])
    return total_cost


def add_items():
# Allows user to add up to three items to the cart
    global user_budget

    if user_budget < 0:
        print("Set your budget first using option 1.")
        return

    display_shop_items()
    print("\nYou may choose up to 3 items.")

    for _ in range(3):
        choice_input = input("Enter item number (1-5): ").strip()

        if choice_input.isdigit():
            choice = int(choice_input)
            item_name, price = fetch_item_details(choice)

            if price == -1:
                print("Invalid selection. Attempt skipped.")
            else:
                if calculate_cart_sum() + price > user_budget:
                    print(
                        "Budget exceeded for",
                        item_name + ". Try cheaper items."
                    )
                else:
                    shopping_cart.append(item_name + " - " + str(price))
                    unique_items.add(item_name.lower())
                    print("Added:", item_name, "| Cost:", price)
        else:
            print("Invalid input. Numbers only. Attempt skipped.")


def show_cart_details():
# Shows all items currently in the cart
    if not shopping_cart:
        print("Your cart has no items.")
        return

    print("\nCURRENT CART:")
    for index in range(len(shopping_cart)):
        print(str(index + 1) + ".", shopping_cart[index])

    total = calculate_cart_sum()
    print("Total cost:", total)

    if user_budget >= 0:
        print("Budget:", user_budget)
        print("Remaining balance:", user_budget - total)


def finalize_purchase():
# Writes purchase details to a receipt file
    global user_budget

    if user_budget < 0:
        print("Please enter a budget before checkout.")
        return

    if not shopping_cart:
        print("Cart is empty. Add items first.")
        return

    total = calculate_cart_sum()
    balance_left = user_budget - total
    receipt_file = "receipt.txt"

    with open(receipt_file, "w") as file:
        file.write("Student Mini-Shop Receipt\n")
        file.write("-------------------------\n")
        file.write("Purchased Items:\n")

        for item in shopping_cart:
            file.write("- " + item + "\n")

        file.write("\nTotal Spent: " + str(total) + "\n")
        file.write("Budget: " + str(user_budget) + "\n")
        file.write("Balance Left: " + str(balance_left) + "\n")

        file.write("\nDifferent Items Purchased:\n")
        for item in unique_items:
            file.write("- " + item + "\n")

    print("\nPurchase completed successfully!")
    print("Receipt saved as:", receipt_file)


while True:
# Program loop keeps running until user exits
    display_menu()
    option_input = input("Choose an option (0-4): ").strip()

    if not option_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    option = int(option_input)

    match option:
        case 1:
            budget_input = input("Enter your budget amount: ")
            if budget_input.isdigit():
                user_budget = int(budget_input)
                print("Budget updated to:", user_budget)
            else:
                print("Invalid amount. Numbers only.")

        case 2:
            add_items()

        case 3:
            show_cart_details()

        case 4:
            finalize_purchase()

        case 0:
            print("Exiting program. Goodbye!")
            break

        case _:
            print("Please select a valid menu option.")

print("\nProgram ended.")
```
