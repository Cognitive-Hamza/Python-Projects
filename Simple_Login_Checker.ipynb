```python
CORRECT_PIN = "658921"
MAX_ATTEMPTS = 3

attempts = 0

while attempts < MAX_ATTEMPTS:
    # Prompt for name (unlimited attempts for name validation)
    while True:
        name = input("Enter your name: ")

        # Validate name (only alphabetic characters)
        if not name.isalpha():
            print("Error: Name must contain only alphabetic characters.")
        else:
            break

    # Prompt for PIN
    pin = input("Enter your PIN: ")

    # Check if PIN is correct
    if pin == CORRECT_PIN:
        print("Correct")
        break
    else:
        print("Error: Incorrect PIN.")
        attempts += 1

# Check if all attempts were used
if attempts >= MAX_ATTEMPTS:
    print("Too many failed attempts")
```
